```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import *
import math
```

## part (c)

```python
def f(x):
    return np.log(np.cosh(x))

def f_p(x):
    return np.sinh(x)/np.cosh(x)

def Q(x,y):
    return f(y)+f_p(y)*(x-y)+(x-y)*(x-y)/2

x = np.linspace(-4, 4, 1000)
fig = plt.figure()
plt.plot(x, f(x),'r', label='f(x)')
plt.plot(x, Q(x,-2),'g', label = 'Q(x,-2)')
plt.plot(x, Q(x,1),'b', label = 'Q(x,1)')
plt.legend(loc='upper left')
plt.show()
```

## part (f)

```python
#update rule
def CalcXn(x):
    x -= np.sinh(x)/np.cosh(x)
    return x
```

```python
# plot xn vs n, x0 = -2
x = -2
step = 10
result = np.ones((step+1,1))
n = np.linspace(0, step, step+1)
result[0,0] = x
for i in range(step):
    x = CalcXn(x)
    result[i+1,0] = x
plt.plot(n,result)
```

```python
# x0 = -2
result
```

```python
# plot xn vs n, x0 = 1
x = 1
step = 10
result = np.ones((step+1,1))
n = np.linspace(0, step, step+1)
result[0,0] = x
for i in range(step):
    x = CalcXn(x)
    result[i+1,0] = x
plt.plot(n,result)
```

```python
# x0 = 1
result
```

## part(g)

```python
#update rule
def CalcXnNewton(x):
    x -= np.sinh(x)*np.cosh(x)
    return x
```

```python
# plot xn vs n, x0 = 1
x = 1
step = 10
result = np.ones((step+1,1))
n = np.linspace(0, step, step+1)
result[0,0] = x
for i in range(step):
    x = CalcXnNewton(x)
    result[i+1,0] = x
plt.plot(n,result)
```

## part(h)

```python
def g(x):
    result = 0
    for k in range(10):
        result += f(x+1/(k+1))
    result /= 10
    return result
```

```python
x = np.linspace(-4, 4, 1000)
fig = plt.figure()
plt.plot(x, g(x),'r', label='g(x)')
plt.legend(loc='upper left')
plt.show()
```

## part(k)

```python
def CalcXnG(x):
    result = 0
    for k in range(10):
        result += f_p(x+1/(k+1))
    result /= 10
    return x - result
```

```python
# plot of xn vs n, x0 = 1
x = 1
step = 20
result = np.ones((step+1,1))
n = np.linspace(0, step, step+1)
result[0,0] = x
for i in range(step):
    x = CalcXnG(x)
    result[i+1,0] = x
plt.plot(n,result)
```

```python
# minimum to 4 significant digits, x0 = 1
result
```
