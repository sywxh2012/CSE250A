```python
import numpy as np
import math
import pandas as pd
import string
```

```python
pw = pd.read_csv('hw1_word_counts_05.txt', sep=" ",header=None)
pw.columns = ['word','count']
pw.head()
```

```python
# Calculate P(w)
pw.loc[:,'P(w)'] = pw['count'].apply(lambda x: x/sum(pw['count']))
pw.sort_values('P(w)', ascending=False).head(15)
```

```python
pw.sort_values('P(w)', ascending=True).head(14)
```

```python
# function to calculate P(E|w)
def p_e_given_w(word,correct,incorrect):
    out = 1
    for i in range(len(correct)):
        if correct[i] in word:
            out = out*1
        else:
            out = out*0
    
    for j in range(len(incorrect)):
        if incorrect[j] in word:
            out = out*0
        else:
            out = out*1
            
    return out    
```

```python
# calculation of P(W|E)
correct = ['A','B']
incorrect = ['C']

pw.loc[:,'P(E|w)'] = pw['word'].apply(lambda x: p_e_given_w(x,correct,incorrect))
```

```python
pw.head()
```

```python
def product(x):
    return x[2]*x[3]
pw.loc[:,'P(E|w)*P(w)'] = pw.apply(product, axis=1)
pw.head()
```

```python
pw.loc[:,'P(w|E)'] = pw['P(E|w)*P(w)'].apply(lambda x: x/sum(pw['P(E|w)*P(w)']))
pw.head()
```

```python
word = pw['word']
list_char = list(string.ascii_uppercase)
pwe = pw['P(w|E)']

def l_in_word(l,word):
    if l in word:
        return 1
    else: 
        return 0

ple = [0] * 26
for ch in list_char:
    i = 0
    ple[i] = 0
    for k in len(word):
        ple[i] += l_in_word(ch,word[k])*pwe[k]
    i += 1
    
```
